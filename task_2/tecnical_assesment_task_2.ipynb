```python
job_title = []
company = []
experience = []
salary = []
location = []
description = []
tags_associated = []
function_area = []
posting_date = [] 
scraping_date = []

exp_sal_loc = []
tag_describe = []

scrape_link = []
unscarpe_link = []
```

```python
for i in links['link']:
    driver = webdriver.Chrome('C:/Users/Nihar/Desktop/TASK/chromedriver_win32/chromedriver.exe')
    driver.get(i)
    soup = bs(driver.page_source)
    driver.close()
    results = soup.find(class_='list')
    if results != None:
        
        job_elems = results.find_all('article',class_='jobTuple')

        for j in job_elems:
            job_title.append(j.find('a',class_='title ellipsis').text)
            company.append(j.find('a',class_='subTitle ellipsis fleft').text)


            for i in j.find('ul'):
                exp_sal_loc.append(i.text)
            experience.append(exp_sal_loc[0])
            salary.append(exp_sal_loc[1])
            location.append(exp_sal_loc[2])

            exp_sal_loc.clear()

            description.append(j.find('div',class_="ellipsis job-description").text)

            for i in j.find('ul', class_="tags has-description"):
                tag_describe.append(i.text)
            fun_area = ' '.join(tag_describe[1:])
            tags_associated.append(tag_describe[0])
            function_area.append(fun_area)

            tag_describe.clear()

            posting_date.append(j.find('span', class_="fleft postedDate").text) 
    else:
        unscarpe_link.append(i)
```

```python
unscarpe_link
```

```python
len(unscarpe_link)
```

```python
scrape_link = []
for i in links['link']:
    if i not in unscarpe_link:
        scrape_link.append(i)
scrape_link
```

```python
df = pd.DataFrame({'job_title':job_title,'company':company,'experience':experience,
                   'salary':salary,'location':location,'description':description,
                   'tags_associated':tags_associated,'function_area':function_area,
                   'posting_date':posting_date})
```
