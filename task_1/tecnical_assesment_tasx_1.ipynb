#### Making synonumus function

```python
import nltk
# download owm-1.4 from nltk
nltk.download('omw-1.4')
```

```python
# download wordnet from nltk
nltk.download('wordnet')
```

```python
#import wordnet libary after downloading
from nltk.corpus import wordnet
```

```python
def synonyms(phrase):
    '''This function take a phrase from user
       and extract all the synomous word
       return set of words to reduce duplicate''' 
    synonyms = []
    for syn in wordnet.synsets(phrase):
        for lm in syn.lemmas():
                 synonyms.append(lm.name())#adding into synonyms
    return set(synonyms)
```

```python
# Extracting synonyms word of high competition and high technolgy
high = synonyms('high')
comp = synonyms('competition')
tech = synonyms('technology')

words = high.union(comp).union(tech)
words
```

```python
# extract those sentense having synonyms word of high competition and high technolgy
for paragraph in extracted_text:
    for word in words:
        if word in paragraph:
            print(paragraph)
```

```python
def extract_paragraph(text,words):
    '''This function extract those senstence 
       having synonymous words'''
    for paragraph in text:
        for word in words:
            if word in paragraph:
                return paragraph
```

# Extracting paragraph from all the html file and store in a text file

```python
sentense_list = []
for file in html_file:
    # load html file
    load_html_file = open(file, "r")
    # read html file
    read_html_file = load_html_file.read()
    # convert html file to text file
    text_file = html2text.html2text(read_html_file)
    # splitting textfile with '\n'
    text_list = text_file.split('\n')
    # append each paragraph with synonyms words
    paragraph = extract_paragraph(text_list,words)
    
    if paragraph == None:
        pass
    else:
        sentense_list.append(paragraph)
```
